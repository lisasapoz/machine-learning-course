## 2
Using this dataset found through Kaggle, I am looking to determine what factors contributed to a passenger surviving the crash and sinking of the Titanic. Data that is known about the passengers includes age, cabin tier (an indicator of socioeconomic status), number of family members onboard (siblings&spouses, parents & children), ticket price, and surival. 

I think this is a particularly interesting dataset as it is important to learn what factors help vs. hinder an individual in surving a disaster.

## 3

I started by merging together a table containing background information about a passenger (age, family member onboard, status) with a table denoting whether a given passenger survived. This helped parse through all relevant data in one tabe. 

The dataset is missing cabin numbers for some first-tier passengers and some of the passenger's ages. The missing cabin numbers don't hinder my ability to draw conclusions from the data as a cabin number is simply a futher indicator of higher socioeconomic status, which is seperately reported. I did not include the passengers missing ages in the age analysis below.



## 4

```python
import pandas as pd
import matplotlib.pyplot as plt

df1=pd.read_csv("test.csv")

plt.scatter(df1.Age, df1.Survived);
plt.xlabel('Passenger Age')
plt.ylabel('Survived or Not')
plt.title('Survival of Passengers by Age')
plt.show();

pclass = df1.Pclass.multiply(df1.Survived)
n, bins, patches = plt.hist(pclass, 3, range = [1,3], facecolor='green', edgecolor="black", linewidth=1, alpha=0.55);
n, bins, patches = plt.hist(df1.Pclass, 3, facecolor='blue', edgecolor="black", linewidth=1, alpha=0.25);
plt.xlabel('Cabin Tier of Passengers')
plt.ylabel('Number of Passengers')
plt.title('Survival of Passengers by Socioeconomic Status')
plt.legend(['Survived', 'Total'])
plt.show();

sibsp = df1.SibSp.where(df1.Survived>0, -1);
n, bins, patches = plt.hist(sibsp, 9, range = [0,9], facecolor='green', edgecolor="black", linewidth=1, alpha=0.55);
n, bins, patches = plt.hist(df1.SibSp, 9, range = [0,9], facecolor='blue', edgecolor="black", linewidth=1, alpha=0.25);
plt.xlabel('Number of Siblings/Spouse Passenger has on Board')
plt.ylabel('Number of Passengers')
plt.title('Survival of Passengers by Number of "Independent" Family Members Onboard');
plt.legend(['Survived', 'Total'])
plt.show();

parch = df1.Parch.where(df1.Survived>0, -1);
n, bins, patches = plt.hist(parch, 10, range = [0,10], facecolor='green', edgecolor="black", linewidth=1, alpha=0.55);
n, bins, patches = plt.hist(df1.Parch, 10, range = [0,10], facecolor='blue', edgecolor="black", linewidth=1, alpha=0.25);
plt.xlabel('Number of Parents/Children on Board')
plt.ylabel('Number of Passengers')
plt.title('Survival of Passengers by Number of "Dependent" Family Members Onboard');
plt.legend(['Survived', 'Total'])
plt.show();

```

From the plot studying age, it is immediately striking how many people died. There were casualties from every age group, no age seems to indicate a significantly higher chance of survival.

It appears that socioeconomic class is the highest predictor of survival rate. Wealthier passengers in first-tier cabins were closer to the deck of the ship were the first to find out about the crash and available lifeboats. News of the crash travelled down to the third-tier passengers last and they were first to experience flooding. 

As I had expected, it also seems beneficial to have a family member on board. In the case of a sibling or spouse, it helped to have someone else looking out for your well-being. Families with children received more compassion and received higher priority for life-boat access. 

## 5
Through analyzing the Titanic passenger records, it is possible to get some insight into what factors enabled some passengers to survive instead of others. The most indicative factor for predicting survival is the socioeconomic status of the passengers. Wealthy passengers were informed of the crash first, had easier access to life boats, and were safe from the flooding on the lower decks. Having a family member on board was also beneficial as there was an extra person looking out for your well-being, and families with children drew greater compassion. Despite this sympathy, deaths are spread out pretty evenly accross all age groups. 

It is important to analyze datasets such as these to learn how humans respond in times of crisis. Who gets priority in a matter of life and death?
